# IMPORTS

```python
import yfinance as yf 
import pandas as pd 
from sklearn.model_selection import train_test_split 
from sklearn.preprocessing import MinMaxScaler 
from sklearn.metrics import mean_squared_error 
#from tensorflow.keras.models import Sequential 
#from tensorflow.keras.layers import Dense, LeakyReLU 
```

```python
# Obter dados históricos da ação BBAS3 dos últimos 12 meses 

ticker = 'CMIN3.SA' 

data = yf.download(ticker, period='1y', interval='1d') 
data
```

# Treat

```python
# Selecionar as colunas relevantes para a análise 
data = data[['Adj Close', 'High', 'Low', 'Volume']].dropna() 
data.columns = ['Close', 'High', 'Low', 'Volume']

# Converter o índice para o formato datetime e definir como índice do DataFrame 
data.index = pd.to_datetime(data.index)

# Criar variáveis para o modelo 
data['Previous_Close'] = data['Close'].shift(1) # Usar o fechamento anterior como preditor 
data['Previous_High'] = data['High'].shift(1) # Usar a máxima anterior como preditor 
data['Previous_Low'] = data['Low'].shift(1) # Usar a mínima anterior como preditor 
data['Previous_Volume'] = data['Volume'].shift(1) # Usar o volume anterior como preditor

# Remover valores nulos resultantes do deslocamento 
data.dropna(inplace=True) 
```

# Model

```python
# Dividir em variáveis independentes (X) e dependentes (y) 

X = data[['Previous_Close', 'Previous_High', 'Previous_Low', 'Previous_Volume']] 

y_high = data['High'] # Previsão da máxima do dia seguinte 

y_low = data['Low'] # Previsão da mínima do dia seguinte 

 

# Dividir em conjuntos de treino e teste 

X_train, X_test, y_high_train, y_high_test = train_test_split(X, y_high, test_size=0.2, random_state=42) 

_, _, y_low_train, y_low_test = train_test_split(X, y_low, test_size=0.2, random_state=42) 

 

# Escalonar os dados para o modelo 

scaler = MinMaxScaler() 

X_train_scaled = scaler.fit_transform(X_train) 

X_test_scaled = scaler.transform(X_test) 
```
