```python
import yfinance as yf
import matplotlib.pyplot as plt

# Funo para gerar o grfico com base nas opes escolhidas
def plot_graph(ticket, period, interval):
    dados = yf.download(ticket, period=period, interval=interval)

    # Remover finais de semana e feriados (dias sem negociao)
    dados = dados.dropna()

    # Criar grfico
    plt.figure(figsize=(10, 6))
    plt.plot(dados.index, dados['Close'], label=f'Cotao de Fechamento ({interval})')

    plt.title(f'Cotao de {ticket} ({period})')
    plt.xlabel('Data')
    plt.ylabel('Preo de Fechamento (R$)')
    plt.xticks(rotation=45)

    # Formatar e ajustar o eixo X para datas
    plt.gca().xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%d-%m-%Y'))  # Formato das datas
    plt.gca().xaxis.set_major_locator(plt.matplotlib.dates.WeekdayLocator())  # Mostrar apenas dias teis (semanas)

    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()

# Solicitar parmetros ao usurio
ticket = input("Digite o ticker da ao (ex: VALE3.SA): ")
period = input("Digite o perodo (opes: 1d, 5d, 1mo, 3mo, 6mo, 1y, 2y, 5y, 10y, ytd, max): ")
interval = input("Digite o intervalo (opes: 1m, 5m, 15m, 30m, 60m, 90m, 1d, 5d, 1wk, 1mo, 3mo): ")

# Gerar o grfico com os parmetros informados
plot_graph(ticket, period, interval)
```
